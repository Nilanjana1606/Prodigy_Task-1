```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Medicaldataset.csv')
df
```

```python
df.head()
```

```python
df.describe()
```

```python
d = {'negative': 0, 'positive':1}
df['Result'] = df['Result'].map(d)
df
```

```python
df.corr()
```

```python
plt.figure(figsize=(10, 6))  


sns.violinplot(x='Result', y='Age', hue='Gender', data=df)

plt.title('Violin Plot')  
plt.xlabel('Result')   
plt.ylabel('Age')     
plt.legend(title='Gender')  

plt.show()  
```

#### Age distributin plot

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.distplot(df['Age'])
```


### blood pressure distribution

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Systolic blood pressure'], shade=True, color='green', label='Systolic BP')
sns.kdeplot(df['Diastolic blood pressure'], shade=True, color='purple', label='Diastolic BP')
plt.title('Blood Pressure Distribution')
plt.xlabel('Blood Pressure')
plt.ylabel('Density')
plt.legend()
plt.show()
```
